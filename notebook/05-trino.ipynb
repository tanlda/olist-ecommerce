```python
import polars as pl
import sqlalchemy as sa
from trino.dbapi import connect
```

```python
connection = connect(host="trino.sql", port=80, user="admin", catalog="database", schema="public")
def execute(query: str): return pl.read_database(query, connection=connection)
```

```python
print(execute("show tables"))
for table in execute("show tables")["Table"]:
    print(table)
    print(execute(f"describe {table}"))
    print(execute(f"select * from {table} limit 2"))
```

```python
document = connect(host="trino.sql", port=80, user="admin", catalog="document", schema="public")
def execute(query: str): return pl.read_database(query, connection=document)
```

```python
print(execute("show schemas from document"))
```

```python
print(execute("show tables from document.ecommerce"))
```

```python
print(execute("describe document.ecommerce.order_reviews"))
```
